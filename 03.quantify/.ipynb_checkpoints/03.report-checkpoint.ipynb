```python
# for featurecount results
# Cell 1: 导入库 + 加载数据 + 清洗样本名
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import re

sns.set(style="whitegrid", font_scale=1.2)
plt.rcParams['figure.figsize'] = (8, 6)

# 读取原始计数文件
df = pd.read_csv("gene_counts.txt", sep="\t", comment="#", index_col=0)

# 提取样本列（第6列之后）
count_matrix = df.iloc[:, 5:].copy()  # 注意：加 .copy() 避免警告

# === 关键：清洗列名 ===
# 原始列名类似：/disk2/cai113/data/stateTrans/02.alignment/L1MKI0600921-w4_1/Aligned.sortedByCoord.out.bam
# 我们想提取：L1MKI0600921-w4_1

new_columns = []
for col in count_matrix.columns:
    # 方法1：用 split('/') 取倒数第二段（目录名）
    parts = col.split('/')
    if len(parts) >= 2:
        sample_name = parts[-2]  # BAM 在 sample_dir/ 下，所以 -2 是目录名
    else:
        # 如果不是路径格式，保留原名
        sample_name = col
    new_columns.append(sample_name)

count_matrix.columns = new_columns

print("计数矩阵形状:", count_matrix.shape)
print("样本名已清洗为:", list(count_matrix.columns))
count_matrix.head()
```

```python
# Cell 2: 读取 summary 并计算 Assigned %
summary = pd.read_csv("gene_counts.txt.summary", sep="\t", index_col=0)

# 转置方便处理
summary_t = summary.T
summary_t['Total Reads'] = summary_t.sum(axis=1)  # 总输入 reads（实际是比对上的）
summary_t['Assigned %'] = summary_t['Assigned'] / summary_t['Total Reads'] * 100

# 打印关键指标
print("Assigned 比例:")
print(summary_t[['Assigned', 'Total Reads', 'Assigned %']])
```

```python
# Cell 3: 样本总 reads 数（柱状图）
total_counts = count_matrix.sum()

plt.figure(figsize=(7, 5))
ax = sns.barplot(x=total_counts.index, y=total_counts.values, palette="Blues_d")
ax.set_title("Total Counts per Sample")
ax.set_ylabel("Total Mapped Reads (to genes)")
plt.xticks(rotation=90, ha='right',fontsize=10)  # ha='right' 让文字右对齐，更紧凑
plt.tight_layout()
plt.show()

print("\n文库大小（总 counts）:")
for sample, total in total_counts.items():
    print(f"  {sample}: {total:,}")
```

```python
# Cell 4: 表达基因数量
expressed_genes = (count_matrix > 0).sum()

plt.figure(figsize=(7, 5))
ax = sns.barplot(x=expressed_genes.index, y=expressed_genes.values, palette="Greens_d")
ax.set_title("Number of Expressed Genes (count > 0)")
ax.set_ylabel("Gene Count")
plt.xticks(rotation=90, ha='right',fontsize=10)
plt.tight_layout()
plt.show()

print("\n表达基因数:")
for sample, n in expressed_genes.items():
    print(f"  {sample}: {n:,} genes")
```

```python
# for samtools results
import pandas as pd
import numpy as np
import glob
import re
import matplotlib.pyplot as plt

# 找到 flagstat 和 idxstats 文件
flagstat_files = sorted(glob.glob("*flagstat.txt"))
idxstats_files = sorted(glob.glob("*idxstats.txt"))

flagstat_files, idxstats_files
```

```python
# 解析 flagstat（mapping rate 等）
def parse_flagstat(file):
    with open(file) as f:
        lines = f.readlines()

    # 提取数字
    def get_num(pattern):
        for line in lines:
            if pattern in line:
                return int(line.split()[0])
        return None

    total = get_num("in total")
    primary = get_num("primary")
    mapped = get_num("mapped (100.00%")
    properly_paired = get_num("properly paired")
    singletons = get_num("singletons")
    
    sample = file.replace(".flagstat.txt", "")

    return {
        "sample": sample,
        "total_reads": total,
        "mapped_reads": mapped,
        "properly_paired": properly_paired,
        "singletons": singletons,
        "mapping_rate": mapped / total * 100 if total else None
    }

flagstat_df = pd.DataFrame([parse_flagstat(f) for f in flagstat_files])
flagstat_df
```

```python
plt.figure(figsize=(10,5))
plt.bar(flagstat_df["sample"], flagstat_df["mapping_rate"])
plt.xticks(rotation=90)
plt.ylabel("Mapping Rate (%)")
plt.title("Mapping Rate per Sample")
plt.show()

plt.figure(figsize=(10,5))
plt.bar(flagstat_df["sample"], flagstat_df["total_reads"])
plt.xticks(rotation=90)
plt.ylabel("Total Reads")
plt.title("Total Reads per Sample")
plt.show()
```

```python
#解析 idxstats

idxstats_files = sorted(glob.glob("*idxstats.txt"))

def parse_idxstats(file):
    df = pd.read_csv(file, sep="\t", header=None,
                     names=["chr", "length", "mapped", "unmapped"])
    sample = file.replace(".idxstats.txt", "")
    df["sample"] = sample
    return df

idxstats_df = pd.concat([parse_idxstats(f) for f in idxstats_files])

# -----------------------------
# 计算每个样本的总 mapped reads
# -----------------------------
total_mapped = idxstats_df.groupby("sample")["mapped"].sum()

# -----------------------------
# 计算线粒体比例 (chrM / total mapped)
# -----------------------------
mt_df = idxstats_df[idxstats_df["chr"].isin(["MT", "chrM"])]

mt_ratio = (mt_df.groupby("sample")["mapped"].sum() / total_mapped) \
            .reset_index().rename(columns={"mapped": "mt_ratio"})

# -----------------------------
# 计算 unplaced (GL / JH / MU) scaffolds 比例
# -----------------------------
unplaced_df = idxstats_df[idxstats_df["chr"].str.contains("GL|JH|MU")]

unplaced_ratio = (unplaced_df.groupby("sample")["mapped"].sum() / total_mapped) \
                 .reset_index().rename(columns={"mapped": "unplaced_ratio"})

# -----------------------------
# 合并输出结果
# -----------------------------
idxstats_summary = total_mapped.reset_index().rename(columns={"mapped": "total_mapped"}) \
                    .merge(mt_ratio, on="sample", how="left") \
                    .merge(unplaced_ratio, on="sample", how="left")

idxstats_summary
```

```python
df = idxstats_summary.copy()

# ------------ 参数设定 ------------
mt_threshold = 0.08   # 线粒体比例阈值
unplaced_threshold = 0.001  # unplaced 阈值（可选）

# 标记需要注意的样本
df["mt_flag"] = df["mt_ratio"] > mt_threshold
df["unplaced_flag"] = df["unplaced_ratio"] > unplaced_threshold

# ------------ 绘图：MT ratio ------------
plt.figure(figsize=(18, 6))
plt.plot(df["sample"], df["mt_ratio"], marker="o")

# 标记超过阈值的点
for i, row in df[df["mt_flag"]].iterrows():
    plt.scatter(row["sample"], row["mt_ratio"], color="red")
    plt.text(row["sample"], row["mt_ratio"], row["sample"], 
             fontsize=8, rotation=45, color="red")

plt.axhline(mt_threshold, color="red", linestyle="--", label=f"MT threshold = {mt_threshold}")

plt.xticks(rotation=80)
plt.ylabel("MT Ratio")
plt.title("MT Ratio per Sample (flagged samples shown)")
plt.tight_layout()
plt.show()

# ------------ 绘图：Unplaced ratio ------------
plt.figure(figsize=(18, 6))
plt.plot(df["sample"], df["unplaced_ratio"], marker="o")

for i, row in df[df["unplaced_flag"]].iterrows():
    plt.scatter(row["sample"], row["unplaced_ratio"], color="red")
    plt.text(row["sample"], row["unplaced_ratio"], row["sample"], 
             fontsize=8, rotation=45, color="red")

plt.axhline(unplaced_threshold, color="red", linestyle="--", 
            label=f"Unplaced threshold = {unplaced_threshold}")

plt.xticks(rotation=80)
plt.ylabel("Unplaced Ratio")
plt.title("Unplaced Ratio per Sample (flagged samples shown)")
plt.tight_layout()
plt.show()
```
